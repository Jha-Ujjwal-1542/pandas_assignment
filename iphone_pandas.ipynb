```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("apple.csv")
data.tail(5)
```

Find the length (number of rows) of dataframe

```python
len(data)
```

Count the number of NaN of each column

```python
nan_count = data.isna().sum()
nan_count
```

Fill NaN of review_text with 0

```python
data['review_text'].fillna(data['review_text'].fillna(0) , inplace=True)
data
```

Fill 0 in place of NaN of review_title

```python
data['review_title'].fillna(data['review_title'].fillna(0) , inplace=True)
data
```

Reset index of dataframe after dropna()

```python
data.reset_index(drop=True, inplace=True)
data
```

Length of dataframe after dropping all NaN

```python
len(data)
```

Get the unique values of a column

```python
unique_products = data['product'].unique()
unique_products
```

```python
data.replace('',value=np.NaN, inplace=True)
data.head()
```

```python
len(data)
```

Get the mean of Total Comments column

```python
data["total_comments"] = data["total_comments"].apply(pd.to_numeric)
data["total_comments"].mean()
```

Get the unique ratings

```python
unique_ratings = data["review_rating"].unique()
unique_ratings
```

Get the dataframe of list according to UniqueRatings   

```python
df_list = []
for x, i in enumerate(unique_ratings):
  state = data[data['review_rating'] == i]
  df_list.append(state)

df_list[1].head(10)
```

Get an array of unique in countries

```python
unique_country = data["review_country"].unique()
unique_country
```

Count frequency of Review Ratings 

```python
df1 = data['review_rating'].value_counts()
df1
```

Count Frequency Value Using GroupBy.size()

```python
df1 = data.groupby(['profile_name', 'reviewed_at']).size()
print(df1)
```
